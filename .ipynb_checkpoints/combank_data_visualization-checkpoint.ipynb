```python
# from google.colab import drive
# drive.mount('/content/drive')

import pandas as pd
df = pd.read_csv('combank.csv')
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a time series plot of closing prices
plt.plot(df['Date'], df['Close'])
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Time Series Plot of Closing Prices')
plt.show()
```

```python
import mplfinance as mpf
import pandas as pd


# Create a candlestick chart of the bank stock data
mpf.plot(df, type='candle', mav=(20,50), volume=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a volume chart of the bank stock data
plt.bar(df['Date'], df['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Chart of Bank Stock Data')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Create a box plot of closing prices by year
df['Year'] = pd.DatetimeIndex(df['Date']).year
sns.boxplot(x='Year', y='Close', data=df)
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.title('Box Plot of Closing Prices by Year')
plt.show()
```
